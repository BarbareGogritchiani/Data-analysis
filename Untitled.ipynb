```python
remote_count = df['RemoteWork'].value_counts().get('Remote', 0)
```

```python
hybrid_count = df['RemoteWork'].value_counts().get('Hybrid (some remote, some in-person)',0)
```

```python
labels = ['Remote', 'Hybrid']
counts = [remote_count, hybrid_count]
```

```python
plt.bar(labels, counts)
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.title('Remote Work Type')
plt.show()
```
